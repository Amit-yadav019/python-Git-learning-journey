```python
import numpy as np 
a = np.random.randint(1,100,20)
print(a)
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
print(b)
```

```python
# np.sort()

np.sort(a)
```

```python
#descending order

np.sort(a)[::-1]
```

```python
np.sort(b) # row waise sorted (default)
```

```python
np.sort(b,axis = 0) #column wise sorting 
```

```python
# np.append()

np.append(a,150)
```

```python
# now  in 2d to add one more column let say (1,1,1,1)
np.append(b,np.ones((b.shape[0],1)),axis = 1 )
```

```python
# np.concatenate()

c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print(d)
```

```python
np.concatenate((c,d),axis = 0)  # concatenate along raw
```

```python
np.concatenate((c,d),axis = 1) # concatenate along column
```
