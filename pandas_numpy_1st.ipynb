```python
import pandas as pd 
import numpy as np 
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# to drop missing values 
df.dropna() # By default, running df.dropna() runs with specific background criteria
```

```python
# Customizing the  dropna() Criteria
df.dropna(axis='index',how='any')
# here any drops any rows that are missing 
```

```python
df.dropna(axis='index',how='all')
# it will drop only rows, where all the elements are misiing 
```

```python
df.dropna(axis='columns',how='all')
# axis='columns' and how='all' .
# as there no any columns have complete missing values so all the columns and rows are there 
```

```python
# and if axis='columns' and how='any' will drop all column containing even a single missing value.
df.dropna(axis='columns',how ='any')
```

# Filtering by Specific Columns (subset)

```python
# If you only want to drop rows missing data in a critical column (e.g., an email address)
df.dropna(axis='index',how='any',subset=['email'])
# here does not any impact that either we use how ='any' or how = 'all', it will give same results 
# as we are checking for a single (one) value 
```

```python
#To check multiple columns where a row is dropped only if both are missing, combine
df.dropna(axis ='index',how='all',subset =['last','email'])
# here all works similar to (AND operation ) , i.e. drop only if both last and email are misssing in a specific row
```

```python
df.dropna(axis ='index',how='any',subset =['last','email'])
# here 'any' works similar to ( OR operations) means it will drop if any of last or email is missing or both.
```
