NOW  USING CONCEPTS OF FILTERING 


```python
df['last'] == 'cena' # to check whether cena is the. last name or not , if the answer is true then is yes , or not is false 
```

```python
filt = (df['last']== 'cena') # created a varible as filt . instead of filter becoz 'filter' is python built in function 
```

```python
df[filt]
```

```python
df.loc[filt,'email'] # now we can get any info. of elements having last = cena .
```

```python
# let if we want first name = john and last name = cena then 
filt1 = (df['last'] == 'cena') & (df['first'] == 'john') # using AND (&) 
df[filt1] 
```

```python
# using OR(|)
filt2 = (df['last'] =='cena') | (df['first'] =='corey')
df[filt2]
```

```python
df.loc[~filt2,'email'] # ~filt2 will gives out ouputs of elements other than which full filled this conditions 
```

apply: Runs a function element-wise on a Series, or axis-wise on a DataFrame.

map: Swaps out specific values in a Series using a lookup dictionary.

applymap: Runs a function on every single individual element across an entire DataFrame.
