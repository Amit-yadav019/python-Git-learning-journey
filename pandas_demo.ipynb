```python
# creating a dictonary

people = {
    "first":["corey","john","harry"],
    "last" : ["schafer","cena","potter"],
    "email":[ "coreyschafer@gmail.com","johncena@gmai;.com","harrypotter@gmail.com"]
             
}
```

```python
people["email"] # to access information within dictonary
```

```python
# creating a dataframe using a dictonary (people)
import pandas as pd 

```

```python
df = pd.DataFrame(people)
df
```

```python
df["email"] # to access information within dataframe ,
# better way to access

# outputs will be series object 
```

```python
type(df["email"])
```

```python
# alternate way to access information within dataframe
df.email
```

```python
# to access multiple information
df[['email','last']] # list ['email','last'] is passed inside this bracket []
# i.e. df[list]. and list = ['email','last']

# NOTE - as we are accessing more than one info. the outputs will now not same as previoulsy as series 
# now the outputs will be dataframe and later we can extract information from them
```

```python
df.columns # to access column info.
```

```python
df.iloc[0] # to access row info having index 0 , i.e. 1st row
# & index here will be column name so we can clearly see what info. represnating what
```

```python
df.iloc[[0,1]]
# to access multiple rows
```

```python
# To access row along with columns to
df.iloc[[0,1],2]

# here  first argument [0, 1] selects the rows, and the second argument 2 selects the third column (index 2)
```

```python
df.loc[0]
```

```python
df.loc[[0,1]]
```
