NOW  USING CONCEPTS OF FILTERING 


```python
df['last'] == 'cena' # to check whether cena is the. last name or not , if the answer is true then is yes , or not is false 
```

```python
filt = (df['last']== 'cena') # created a varible as filt . instead of filter becoz 'filter' is python built in function 
```

```python
df[filt]
```

```python
df.loc[filt,'email'] # now we can get any info. of elements having last = cena .
```

```python
# let if we want first name = john and last name = cena then 
filt1 = (df['last'] == 'cena') & (df['first'] == 'john') # using AND (&) 
df[filt1] 
```

```python
# using OR(|)
filt2 = (df['last'] =='cena') | (df['first'] =='corey')
df[filt2]
```

```python
df.loc[~filt2,'email'] # ~filt2 will gives out ouputs of elements other than which full filled this conditions 
```

LECTURE 05 - Updating Rows & Columns - modifying data within dataframes 

```python
df.columns
```

```python
df.columns = ['first_name','last_name','email']
df.columns
```

```python
# to convert coulumns into uppercase 
df.columns = [x.upper() for x in df.columns]
df
```

```python
# to replaces spaces with _ . for example if there is a coloumn name first name , then df. first name will show and error message 
# so we need to add _ b/w them for example first_name
df.columns = df.columns.str.replace(' ','_') # it will convert spaces into _
df
```

```python
# what if  we we wanted to do some changes in specific columns 
# let we wanted to change first_name and last_name as first and last
df.rename(columns={'FIRST_NAME': 'first' , 'LAST_NAME':'last'}, inplace=True)
df
```

```python
# UPDATING DATA IN A ROWS 
df.loc[3] = ['Roman','reings','romanreings@gmail.com']
df
```

```python
# lets again we wanted to change the name roman reings as roman cena 
df.loc[3,[ 'last','EMAIL']]  = ['cena','romancena@gmail.com']
df
```

```python
df.loc[2,'last'] = 'wonder'
df
```

```python
# Alternate method by using at
df.at[2,'last'] = 'potter'
df
```

```python
filt = (df['EMAIL'] == 'harrypotter@gmail.com')
df[filt]
```

```python
# IS WE WANT TO ACESSS ONLY LAST NAME THEN WE CAN DO LIKE 
filt = (df['EMAIL'] == 'harrypotter@gmail.com')
df[filt]['last']
```

```python
# trying to update single row of data 
filt = (df['EMAIL'] == 'harrypotter@gmail.com')
df[filt]['last'] = ['kane']
```

```python
# so to fix the above error we have to use loc function 
filt = (df['EMAIL'] == 'harrypotter@gmail.com')
df.loc[filt,'last'] = ['kane']
df
```

```python
# updating multiple row of data 
# let convert all tge emails to lowercase 

df['EMAIL'] = df['EMAIL'].str.lower()
df
```

apply: Runs a function element-wise on a Series, or axis-wise on a DataFrame.

map: Swaps out specific values in a Series using a lookup dictionary.

applymap: Runs a function on every single individual element across an entire DataFrame.

replace: Swaps values in a Series without accidentally wiping out unmapped fields.
