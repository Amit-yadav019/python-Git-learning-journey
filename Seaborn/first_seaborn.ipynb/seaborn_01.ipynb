```python
# categoricalplot = catplot
plt.style.use('fivethirtyeight')
sns.catplot(x ='day',y='tip',kind='strip',data=data)
# here kind = 'strip' is by default 




# as we can see there for example at sunday and tip = 4 , here we see 3 ponts having same tip value = 4 , but it does not overlap
# over each other , this is becoz the help of a junction called #jitter 
# so by default jitter is running along with catplot .
```

If multiple observations share identical values, they blend into one another. To fix this, Seaborn uses an internal horizontal adjusting noise parameter called jitter. #jitter

```python
# what if jitter = 0  or jitter = False

sns.catplot(x ='day',y='tip',kind='strip',jitter= 0,data=data)

# by default it is kind ="strip"

# as jitter = 0, there is no any horizontal movement of points so all identical points seems overlapping over each other . 

# we can also increase the values of jitter , the horizontal spread also increase . 
```

# a). Heatmaps

a heat map is a graphical representation of data where the individuals values contained in a matrix are represented as colors.


generally drawn b/w two categorical data or a discrite numerical data .

```python
data_1 = pd.read_csv('flights.csv')
data_1.head()
```

```python

# A heatmap cannot take a standard flat DataFrame directly as an input. 
# To plot it successfully, you must transform your data into a pivot table / cross-tabulation table where:

# The rows represent one categorical variable (e.g., Years).

# The columns represent a second categorical variable (e.g., Months).

# The intersecting cells contain aggregated numerical values (e.g., Passenger counts
```

```python
# Preparing Data with Pandas (pivot_table)
x = data_1.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')

```

```python
print(x)
```

```python
sns.heatmap(x)
```

```python
# we can also increase the distance between two adjusent shells 
# by using linewidth parameter 
sns.heatmap(x,linewidths=0.5)
```

```python
# we can also fill the value in each of the shell by using annotation parameter , i.e. annot = True 
sns.heatmap(x,linewidths=0.5,annot=True)
```

```python
sns.heatmap(x,linewidths=0.5,annot=True,fmt='d')

# using parameter fmt ='d' , so that output will show only integer values 
```

```python
# to change color tone by passing parameter called cmap= 'viridis' or 'summer' or 'winter'
sns.heatmap(x,linewidths=0.5,annot=True,fmt='d',cmap='viridis')
```

# Clustermap

A clustermap is an advanced hierarchical data visualization tool built directly on top of a heatmap.

While a standard heatmap displays data in its raw, chronological, or alphabetical row-and-column order, a clustermap automatically groups and rearranges rows and columns based on their statistical similarity.

```python
x = data_1.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
import scipy
print(scipy.__version__)
```

```python
sns.clustermap(x)
```

Output Interpretation: Observe the horizontal X-axis labels. The months are no longer displayed in their default chronological order (January, February, March...). Instead, high-volume summer vacation months like July and August are grouped right next to each other because their passenger density footprints match almost perfectly.

```python
# simmilarly if we wanted to do for rows only then use , column_cluster = False
sns.clustermap(x,z_score= 0, col_cluster=False,annot = True)
```
