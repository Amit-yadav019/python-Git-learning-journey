```python
#Broadcasting 
a = np.arange(12).reshape(4,3)
b = np.arange(3)
print(a)
print(b)
```

```python
print(a.shape)
print(b.shape)
print(a)
print(b)
print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(3)
print(a)
print(b)
# print(a+b) will not work as by following the rule of broadcasting a = (3,4) and b will become b = (3,3).
#              and by rule error will rasie as there is dimenstional error , 
#     i.e. dimention of a = (3,4) will not be equal to b = (3,3)
```

```python
a = np.arange(4).reshape(1,4)
b = np.arange(4).reshape(4,1)
print(a)
print(b)
print(a+b) # here both a and b will become (4,4)
```

```python
a = np.arange(4).reshape(2,2)
b = np.arange(1).reshape(1)
print(a)
print(b)
print(a+b) # sequence of b will be b= (1,1) - b=(2,1) - b= (2,2) then both a = (2,2 ) and b =(2,2)
```
