# Multi-Column Custom Aggregation

Running a flat aggregation method like .mean() across your entire DataFrame during a resample operation  
can give you logically incorrect data. For example, calculating the mathematical average of a High or Low column 
does not show the true absolute high or low price for that period. Similarly, you typically want the 
total sum of Volume traded, not the average volume.

To apply different aggregation methods to different columns at the same time, pass a dictionary mapping 
your column names to their chosen functions into the .agg() method.
